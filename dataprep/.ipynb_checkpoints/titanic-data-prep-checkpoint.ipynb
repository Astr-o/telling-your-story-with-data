# Titanic Data Prep

Goal is to clean up the titanic dataset to be able to use it as a teaching aid for Kizan.

## Load Data

```python
import pandas as pd
import numpy as np
import pandas_profiling
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
raw_data = pd.read_csv('../datasets/raw_titanic_data.csv')
raw_data.head()
```

```python
pandas_profiling.ProfileReport(raw_data)
```

```python
clean_data = raw_data.copy()

clean_data = clean_data.drop(['Cabin', 'Ticket'], axis=1)
clean_data.head()
```

```python
clean_data.Fare.plot.hist(bins=100)
```

```python
clean_data[clean_data.Fare == 0]
```

Zero fares make sence here as some passengers may have been given a free ticket

```python
clean_data.Age.plot.hist()
```

Try mean, median imputation

```python
from sklearn.preprocessing import Imputer
```

```python
print('null ages')
clean_data.Age.isnull().sum()
```

```python
mean_imp = Imputer(missing_values='NaN', strategy='mean', axis = 0)
median_imp = Imputer(missing_values='NaN', strategy='median', axis = 0)
```

```python
mean_imp_result = clean_data.copy()

mean_imp_result['Age'] = mean_imp.fit_transform(mean_imp_result[['Age']])
mean_imp_result.columns = clean_data.columns
mean_imp_result.index = clean_data.index

mean_imp_result.head()
```

```python
median_imp_result = clean_data.copy()

median_imp_result['Age'] = median_imp.fit_transform(median_imp_result[['Age']])
median_imp_result.columns = clean_data.columns
median_imp_result.index = clean_data.index

median_imp_result.head()
```

```python
clean_data.Age.plot.hist()
plt.show()
mean_imp_result.Age.plot.hist()
plt.show()
median_imp_result.Age.plot.hist()
```

Messes with the distrubution

best options is probably to drop nulls

```python
clean_data = clean_data[~clean_data.Age.isnull()]
clean_data = clean_data[~clean_data.Embarked.isnull()]
clean_data.head()
```

```python
clean_data.shape
```

## Rename Cols

```python
clean_data['Passenger_class'] = clean_data['Pclass']
clean_data['Number_of_siblings_and_spouses'] = clean_data['SibSp']
clean_data['Number_of_parents_and_childern'] = clean_data['Parch']
clean_data['Embarked'] = clean_data['Embarked'].map({'C': 'Cherbourg', 'Q': 'Queenstown', 'S': 'Southhampton'})

clean_data = clean_data.drop(['Pclass', 'SibSp', 'Parch', 'Embarked'], axis=1)
clean_data.head()
```

final check for nulls and cardinality

```python
clean_data.isnull().sum()
```

## Externalise Clean Data

```python
clean_data.to_csv('../datasets/clean_titanic.csv', index_label='index')

test = pd.read_csv('../datasets/clean_titanic.csv', index_col='index')
test.head()
```
